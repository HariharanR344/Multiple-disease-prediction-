# **Parkinsons Prediction**


```python
import pandas as pd
```

```python
parkinsons=pd.read_csv("https://raw.githubusercontent.com/HariharanR344/Multiple-disease-prediction-/refs/heads/main/parkinsons%20-%20parkinsons.csv")
parkinsons
```

# **Checking for Null values**

```python
parkinsons.isnull().sum()
```

```python
parkinsons.dtypes
```

# **Feature and Target Selection**

```python
parkinsons_features=parkinsons[['PPE', 'spread1', 'spread2', 'D2', 'RPDE', 'DFA', 'HNR', 'MDVP:Jitter(%)', 'MDVP:Shimmer', 'NHR']]
parkinsons_target=parkinsons['status']
```

```python
X = parkinsons_features
y = parkinsons_target
```

# **Spliting**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=4, stratify=y
)
```

# **Imbalanced Datset**

```python
parkinsons_count=y.value_counts()
parkinsons_count
```

# **Under Sampling**

```python
! pip install -U imbalanced-learn
```

```python
from imblearn.under_sampling import RandomUnderSampler


rus = RandomUnderSampler()
X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)
```

```python
parkinsons_count=y_train_rus.value_counts()
parkinsons_count
```

```python
ax = y_train_rus.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title("Under-sampling")
```

# **Scaling**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train_rus)
X_test_scaled = scaler.transform(X_test)
```

# **Training**

```python
from sklearn.linear_model import LogisticRegression

parkinsons_model = LogisticRegression()
parkinsons_model.fit(X_train_scaled, y_train_rus)
```

# **Prediction**

```python
ypredicted=parkinsons_model.predict(X_test_scaled)
print(ypredicted)
```

# **Classification Report**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, ypredicted))
```

# **Kidney Disease Prediction**

```python
import gradio as gr
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

# Load models
with open("parkinsons_model.pkl", "rb") as f:
    parkinsons_model = pickle.load(f)
with open("kidney_model.pkl", "rb") as f:
    kidney_model = pickle.load(f)
with open("liver_model.pkl", "rb") as f:
    liver_model = pickle.load(f)

# --- Re-fitting scalers for Parkinsons and Kidney ---
# This is necessary because only the last scaler (for liver) was saved.
# In a production scenario, all scalers should be saved individually.

# Parkinson's data and scaler
parkinsons_df = pd.read_csv("https://raw.githubusercontent.com/HariharanR344/Multiple-disease-prediction-/refs/heads/main/parkinsons%20-%20parkinsons.csv")
parkinsons_features_cols = ['PPE', 'spread1', 'spread2', 'D2', 'RPDE', 'DFA', 'HNR', 'MDVP:Jitter(%)', 'MDVP:Shimmer', 'NHR']
X = parkinsons_df[parkinsons_features_cols]
y = parkinsons_df['status']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4, stratify=y)
rus_parkinsons = RandomUnderSampler()
X_train_rus, y_train_rus = rus_parkinsons.fit_resample(X_train, y_train)
parkinsons_scaler = MinMaxScaler()
parkinsons_scaler.fit(X_train_rus)

# Kidney data and scaler
kidney_df = pd.read_csv("https://raw.githubusercontent.com/HariharanR344/Multiple-disease-prediction-/refs/heads/main/kidney_disease%20-%20kidney_disease.csv")
# Preprocessing for kidney data (as done in the notebook)
cols_to_convert_to_numeric = ['pcv', 'wc', 'rc']
for col in cols_to_convert_to_numeric:
    kidney_df[col] = kidney_df[col].replace('?', np.nan)
    kidney_df[col] = pd.to_numeric(kidney_df[col], errors='coerce')
num_cols = kidney_df.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    kidney_df[col] = kidney_df[col].fillna(kidney_df[col].median())
cat_cols = kidney_df.select_dtypes(include='object').columns
for col in cat_cols:
    kidney_df[col] = kidney_df[col].fillna(kidney_df[col].mode()[0])

le = LabelEncoder()
kidney_df['rbc_encoded']=le.fit_transform(kidney_df['rbc'])
kidney_df['htn_encoded']=le.fit_transform(kidney_df['htn'])
kidney_df['dm_encoded']=le.fit_transform(kidney_df['dm'])
kidney_df['classification_encoded']=le.fit_transform(kidney_df['classification'])

kidney_feature_cols = ['sc', 'bu', 'al', 'hemo', 'pcv', 'sg', 'rbc_encoded', 'htn_encoded', 'dm_encoded', 'bgr']
A = kidney_df[kidney_feature_cols]
B = kidney_df['classification_encoded']
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.2, random_state=4, stratify=B)
rus_kidney = RandomUnderSampler()
A_train_rus, B_train_rus = rus_kidney.fit_resample(A_train, B_train)
kidney_scaler = MinMaxScaler()
kidney_scaler.fit(A_train_rus)

# Liver scaler (this one was correctly saved)
with open("scaler.pkl", "rb") as f:
    liver_scaler = pickle.load(f)

# --- Prediction Functions ---

def predict_parkinsons(
    PPE, spread1, spread2, D2, RPDE, DFA, HNR, MDVP_Jitter_percent, MDVP_Shimmer, NHR
):
    features = np.array([
        PPE, spread1, spread2, D2, RPDE, DFA, HNR, MDVP_Jitter_percent, MDVP_Shimmer, NHR
    ]).reshape(1, -1)
    scaled_features = parkinsons_scaler.transform(features)
    prediction = parkinsons_model.predict(scaled_features)[0]
    return "Parkinson's Disease Detected" if prediction == 1 else "No Parkinson's Disease"

def predict_kidney(
    sc, bu, al, hemo, pcv, sg, rbc, htn, dm, bgr
):
    # Map dropdown values back to encoded integers
    rbc_encoded = 1 if rbc == 'normal' else 0
    htn_encoded = 1 if htn == 'yes' else 0
    dm_encoded = 1 if dm == 'yes' else 0

    features = np.array([
        sc, bu, al, hemo, pcv, sg, rbc_encoded, htn_encoded, dm_encoded, bgr
    ]).reshape(1, -1)
    scaled_features = kidney_scaler.transform(features)
    prediction = kidney_model.predict(scaled_features)[0]
    return "Kidney Disease Detected" if prediction == 0 else "No Kidney Disease"

def predict_liver(
    Age, Gender, Total_Bilirubin, Direct_Bilirubin, Alkaline_Phosphotase,
    Alamine_Aminotransferase, Aspartate_Aminotransferase, Total_Protiens,
    Albumin, Albumin_and_Globulin_Ratio
):
    gender_encoded = 1 if Gender == 'Male' else 0

    features = np.array([
        Age, gender_encoded, Total_Bilirubin, Direct_Bilirubin,
        Alkaline_Phosphotase, Alamine_Aminotransferase, Aspartate_Aminotransferase,
        Total_Protiens, Albumin, Albumin_and_Globulin_Ratio
    ]).reshape(1, -1)
    scaled_features = liver_scaler.transform(features)
    prediction = liver_model.predict(scaled_features)[0]
    return "Liver Disease Detected" if prediction == 1 else "No Liver Disease"

# --- Gradio Interfaces ---

# Parkinson's Interface
parkinsons_inputs = [
    gr.Number(label='PPE'),
    gr.Number(label='spread1'),
    gr.Number(label='spread2'),
    gr.Number(label='D2'),
    gr.Number(label='RPDE'),
    gr.Number(label='DFA'),
    gr.Number(label='HNR'),
    gr.Number(label='MDVP:Jitter(%)'),
    gr.Number(label='MDVP:Shimmer'),
    gr.Number(label='NHR')
]
parkinsons_output = gr.Label()
parkinsons_interface = gr.Interface(
    fn=predict_parkinsons,
    inputs=parkinsons_inputs,
    outputs=parkinsons_output,
    title="Parkinson's Disease Prediction",
    description="Enter the patient's voice feature values to predict Parkinson's disease."
)

# Kidney Disease Interface
kidney_inputs = [
    gr.Number(label='Serum Creatinine (sc)'),
    gr.Number(label='Blood Urea (bu)'),
    gr.Number(label='Albumin (al)'),
    gr.Number(label='Hemoglobin (hemo)'),
    gr.Number(label='Packed Cell Volume (pcv)'),
    gr.Number(label='Specific Gravity (sg)'),
    gr.Radio(['normal', 'abnormal'], label='Red Blood Cells (rbc)'),
    gr.Radio(['yes', 'no'], label='Hypertension (htn)'),
    gr.Radio(['yes', 'no'], label='Diabetes Mellitus (dm)'),
    gr.Number(label='Blood Glucose Random (bgr)')
]
kidney_output = gr.Label()
kidney_interface = gr.Interface(
    fn=predict_kidney,
    inputs=kidney_inputs,
    outputs=kidney_output,
    title="Kidney Disease Prediction",
    description="Enter patient's details to predict Kidney disease."
)

# Liver Disease Interface
liver_inputs = [
    gr.Number(label='Age'),
    gr.Radio(['Male', 'Female'], label='Gender'),
    gr.Number(label='Total Bilirubin'),
    gr.Number(label='Direct Bilirubin'),
    gr.Number(label='Alkaline Phosphotase'),
    gr.Number(label='Alamine Aminotransferase'),
    gr.Number(label='Aspartate Aminotransferase'),
    gr.Number(label='Total Protiens'),
    gr.Number(label='Albumin'),
    gr.Number(label='Albumin and Globulin Ratio')
]
liver_output = gr.Label()
liver_interface = gr.Interface(
    fn=predict_liver,
    inputs=liver_inputs,
    outputs=liver_output,
    title="Liver Disease Prediction",
    description="Enter patient's details to predict Liver disease."
)

# --- Combine interfaces into a Tabbed Interface ---
demo = gr.TabbedInterface([
    parkinsons_interface,
    kidney_interface,
    liver_interface
], [
    "Parkinson's Prediction",
    "Kidney Disease Prediction",
    "Liver Disease Prediction"
])

demo.launch(debug=True)
```
